# Datapack Sandbox：可执行 MCF 文档

这个 Notebook 使用持久沙盒运行原生 `.mcfunction` 单元，并在每次执行后显示状态变化和 PNG。

```mcfunction
%dps version 26.2
%dps config autoRender true
%dps reset --apply
%dps status
```

## 创建可观察状态

下面的命令创建计分板、放置方块并生成实体。它们都保留在同一个 Kernel 世界中。

```mcfunction
scoreboard objectives add notebook dummy
scoreboard players set Steve notebook 1
setblock 0 0 2 minecraft:stone
setblock 1 0 2 minecraft:glass
summon minecraft:zombie 2 0 4
```

## 继续修改同一世界

```mcfunction
scoreboard players add Steve notebook 4
time set day
say Notebook state updated
```

使用本机原版资源时，在新的控制单元中运行 `%dps assets <client.jar>`，然后运行 `%dps render`。
